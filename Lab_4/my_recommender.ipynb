```python
latentusers = user_prefs.iloc[:,1:]
print(latentusers.shape)
latentitems = items.iloc[:,2:]
latentitems.shape
# latent_user = np.array()
```

```python
# np.array(latentusers.values)
# latentitems.values
```

```python
latent_user_preferences = np.array(latentusers.values)
print(latent_user_preferences.shape)
latent_item_features = np.array(latentitems.values)
print(latent_item_features.T.shape)
```

```python
# print(user_prefs[user_prefs['User'] == 1882].index)
```

```python
def predict_rating(user_id, item_id):
    """ Predict a rating given a user_id and an item_id.
    """
    user_preference = latent_user_preferences[user_prefs.index[user_prefs['User'] == user_id]]
    item_preference = latent_item_features[items.index[items['Movie ID'] == item_id]]
    return user_preference.dot(item_preference.T)
```

```python
print(predict_rating(156,98))
print(predict_rating(3806,603))
```
